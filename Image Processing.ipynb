```python
img = cv2.imread('messi5.jpg',0)
plt.imshow(img)
plt.show()

ret, thresh1 = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY_INV)

plt.imshow(thresh1,cmap=plt.cm.Greys)
plt.show()

plt.imshow(thresh2,cmap=plt.cm.Greys)
plt.show()
```

```python
img = cv2.medianBlur(img, 5)
res, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
thresh2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                              cv2.THRESH_BINARY, 11, 2)
thresh3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, 
                                cv2.THRESH_BINARY, 11, 2)
images = [img, thresh1, thresh2, thresh3]

for i in range(4):
    plt.subplot(2,2,i+1), plt.imshow(images[i], cmap='gray')
plt.show()
```

```python
rows, cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(dst)
plt.show()
```

```python
rows, cols = img.shape

M = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1)
dst = cv2.warpAffine(img, M, (cols,rows))

plt.imshow(dst)
plt.show()
```
